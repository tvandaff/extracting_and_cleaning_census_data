```python
#First input the required programs for this project
import requests 
from bs4 import BeautifulSoup
from lxml import html
import re
import pandas as pd
```

```python
#I use URL to define the current estimates website, requests.get() to retrieve the website, and Beautiful Soup
#to parse the website code into readable html format.
url = 'https://www.census.gov/programs-surveys/popest.html'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
#Tbe following code saves the website 
webcontent = str(BeautifulSoup(response.content, 'html.parser'))
f = open('census.html', 'w', encoding="utf-8")
f.write(webcontent)
f.close
```

```python
#The following creates a new variable and places all of the weblinks into the list. I used the reference: 
#https://stackoverflow.com/questions/3075550/how-can-i-get-href-links-from-html-using-python to identify the proper formatting 
#to extract urls 
bsurls = []
for item in soup.find_all('a'):
    bsurls.append(item.get('href'))
    print(bsurls)
```

```python
#From the above output, it is clear that when scanning the data, for sections without a link, the code has input None
#To remove this oversite and an odd outlier in the series, I created the following code: 
valurls=[]
for url in bsurls:
    if (url != None) & (url != '#skipsideNav'):
        valurls.append(url)
valurls
```

```python
#From the above code, there are several links that are missing the first part of the address to the census bureau webpage
#To append those weblinks, I developed the following: 
httpurls = []
for url in valurls: 
    if url.startswith('https://'):
        httpurls.append(url)
    else: 
        url = 'https://www.census.gov' + url
        httpurls.append(url)
httpurls
```

```python
#This code creates two series with the inbound and outbound urls. Inbound urls travel to different pages within the census
#bureau site. The outbound urls travel to webpages outside of the census bureau site. 
#I used the reference: https://stackoverflow.com/questions/34775295/python-startswith-for-loop-and-if-statement, to find the 
#proper function that matches the beginning of a string
outbound = []
inbound = []
for url in httpurls: 
    if url.startswith('https://www.census'):
        inbound.append(url)
    else: 
        outbound.append(url)
```

```python
#I then develop a code to transform the above inbound series into a dataframe for readibility and attached a column to 
#designate the type as inbound
df1 = pd.DataFrame(inbound, columns=['urls'])
df1['type'] = 'inbound'
df1
```

```python
#I then develop a code to transform the above outbound series into a dataframe for readibility and attached a column to 
#designate the type as outbound
df2 = pd.DataFrame(outbound, columns=['urls'])
df2['type'] = 'outbound'
df2
```

```python
#Finally, I concatenated the dataframes to form a single dataframe with a type designator to indicate whether a url was 
#inbound or outbound
frames = [df1, df2]
df = pd.concat(frames)
df
```

```python
df = df.drop_duplicates(subset=['urls'], keep='first')
df
```

```python
#Test to ensure duplicates are removed
df[df.duplicated(subset=['urls'])]
```

```python
#The last necessary act is to clean up the index to avoid duplicate and out-of-order numbering: 
# This was inspired by the stackoverflow post: https://stackoverflow.com/questions/14487562/drop-row-in-pandas-series-and-clean-up-index
df = df.reset_index(drop=True)
df
```

```python
#To save the dataframe into a csv file  use to_csv: 
df.to_csv('page_urls.csv')
```
